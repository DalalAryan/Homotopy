```julia
sum = 3 + 7
println("Sum is ", sum)

difference = 10 - 4
println("Difference is ", difference)

product = 6 * 9
println("Product is ", product)

quotient = 100 / 10
println("Quotient is ", quotient)

power = 10 ^ 2
println("Power is ", power)

modulus = 101 % 2
println("Modulus is ", modulus)
```

```julia
days = 365
days_float = convert(AbstractFloat, days)
println("Days as float: ", days_float)

@assert days == 365 # check if days is 365
@assert days_float == 365.0 # check if days_float is 365.0
```
